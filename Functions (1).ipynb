6. Explain the concept of generators in Python and how they are defined

Generators are a special type of iterator in Python that allow you to iterate over a sequence of values without storing the entire sequence in memory. They are defined using functions with the `yield` keyword.

When a generator function is called, it returns a generator object without starting execution immediately. The execution starts when the `next()` method is called on the generator object or when it's used in a loop. The `yield` statement pauses the function execution and returns a value to the caller. When `next()` is called again, the function resumes from where it left off.

Example:

def count_up(n):
    for i in range(n):
        yield i


7. What are the advantages of using generators over regular functions?

*   **Memory Efficiency:** Generators produce items one at a time as needed, rather than creating the entire sequence in memory at once. This is especially beneficial when dealing with very large sequences that would otherwise consume a significant amount of memory.
*   **Lazy Evaluation:** Values are generated on the fly, only when requested (e.g., by `next()` or a loop). This means that computation is only performed for the items you actually consume, which can improve performance if you don't need the entire sequence.
*   **Infinite Sequences:** Generators can represent infinite sequences because they don't need to store all elements. They can generate values indefinitely as long as the `yield` statement is reached.
*   **Readability and Simplicity:** For certain types of iterative tasks, generators can make the code more concise and easier to read compared to building and returning a list.
*   **Pipelining:** Generators can be easily chained together to create efficient data processing pipelines, where the output of one generator becomes the input of another.

9. Explain the purpose and usage of the `map()` function in Python.

The `map()` function in Python is a built-in function that applies a given function to each item of an iterable (like a list or tuple) and returns an iterator that yields the results. It's a concise way to perform the same operation on all elements of a sequence.

**Purpose:**

*   To apply a transformation to every item in an iterable without using an explicit loop.
*   To make code more readable and often more performant for simple transformations.

**Usage:**

The `map()` function has the following syntax:

````python
# Example 1: Squaring numbers in a list
numbers = [1, 2, 3, 4, 5]

# Using a defined function
def square(x):
  return x * x

squared_numbers_map = map(square, numbers)
print(list(squared_numbers_map)) # Convert the map object to a list to see the results

# Using a lambda function (more common with map)
squared_numbers_lambda_map = map(lambda x: x * x, numbers)
print(list(squared_numbers_lambda_map))

# Example 2: Applying a function with multiple iterables
list1 = [1, 2, 3]
list2 = [4, 5, 6]

def add(x, y):
  return x + y

sum_list = map(add, list1, list2)
print(list(sum_list))
````

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python.

The `map()`, `reduce()`, and `filter()` functions are all built-in tools in Python for working with iterables, but they serve different purposes:

*   **`map()`**: Applies a function to each item in an iterable and returns an iterator of the results. It's used for transforming each element individually.
*   **`filter()`**: Creates an iterator from elements of an iterable for which a function returns true. It's used for selecting elements based on a condition.
*   **`reduce()`**: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It's used for aggregating or combining elements.

Here's a more detailed look at each with examples:

````python
# Example with map()
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(f"Original list: {numbers}")
print(f"After map (squaring): {squared}")
````

````python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
````

````python
 #5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration
````

````python
# Example
fibonacci = FibonacciIterator(50)
for number in fibonacci:
    print(number, end=' ')
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

example = powers_of_two(5)
for power in example:
    print(power, end=' ')
````
